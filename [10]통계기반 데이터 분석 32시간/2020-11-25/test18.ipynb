```python
# 사용할 패키지 정의
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical

# Train, Test split
dataset = np.loadtxt("haberman.data", delimiter=",")
dataset.shape
x_train = dataset[:200, :3]
y_train = to_categorical(dataset[:200:, 3])
x_test = dataset[200:, :3]
y_test = to_categorical(dataset[200:, 3])
y_train = y_train[:,1:3]
y_test = y_test[:,1:3]
# dataset.shape, x_train.shape, y_train.shape, x_test.shape, 
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=3))
model.add(Dense(units=200, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
model.fit(x_train, y_train, epochs=1000, batch_size=3)

```
