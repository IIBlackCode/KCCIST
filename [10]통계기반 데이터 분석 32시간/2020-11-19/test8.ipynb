```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("x / y test graphics")
plt.show()
```

```python
x = np.linspace(-3, 3, 10)
y = x ** 2
plt.plot(x, y, 'r')
plt.plot(x, np.sin(x), 'b')
plt.plot(x, np.cos(x), 'g--')
plt.show()
```
